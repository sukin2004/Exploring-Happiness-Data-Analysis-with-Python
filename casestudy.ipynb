```python
sns.boxplot(data=df, x='continent', y='score')
plt.xticks(rotation=90)
plt.show()
```

```python



corr = df.corr(numeric_only=True)
print(corr['score'])

plt.imshow(corr, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.title("Correlation Matrix")
plt.show()
```

```python
continent_summary = df.groupby('continent')['score'].mean()

continent_summary.plot(kind='bar', figsize=(7,5), title="Average Score by Continent")
plt.ylabel("Average Score")
plt.show()
```

```python
import plotly.express as px

fig = px.choropleth(df,
                    locations='country or region',
                    locationmode='country names',
                    color='score',
                    hover_name='country or region',
                    title="World Happiness Score")
fig.show()
```

```python
df.plot(kind='scatter', x='healthy life expectancy', y='score', figsize=(7,5), color='brown')
plt.title("Healthy Life Expectancy vs Score")
plt.show()

df.plot(kind='scatter', x='social support', y='score', figsize=(7,5), color='blue')
plt.title("Social Support vs Score")
plt.show()
```

```python
df['continent'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Country Share by Continent")
plt.ylabel("")
plt.show()
```
